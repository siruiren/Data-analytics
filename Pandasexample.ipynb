```python
import pandas as pd
import numpy as np
```

```python
# create some random dataset with columns a, b, c
left = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                     'key2': ['K0', 'K1', 'K0', 'K1'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})


right = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                      'key2': ['K0', 'K0', 'K0', 'K0'],
                      'C': ['C0', 'C1', 'C2', 'C3'],
                      'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
left
```

```python
right
```

```python
# inner join
inner_join = pd.merge(left, right, how='inner', on=['key1', 'key2'])
inner_join
```

```python
#outer join
outer_join = pd.merge(left, right, how='outer', on=['key1', 'key2'])
outer_join
```

```python
#right_join
right_join = pd.merge(left, right, how='right', on=['key1', 'key2'])
right
```

```python
#left join
left_join = pd.merge(left, right, how='left', on=['key1', 'key2'])
left_join
```

```python
# concatenate()
# concat by rows (axis 0)
concat_by_axis_0 = pd.concat([left, right], axis =0)
concat_by_axis_0
```

```python
# concat by rows (axis 1)
concat_by_axis_1 = pd.concat([left, right], axis =1)
concat_by_axis_1
```

```python
#combine_first
df1 = pd.DataFrame([[np.nan, 3., 5.], [-4.6, np.nan, np.nan],
                    [np.nan, 7., np.nan]])

df2 = pd.DataFrame([[-42.6, np.nan, -8.2], [-5., 1.6, 4]],
                   index=[1, 2])
```

```python
df1
```

```python
df2
```

```python
result = df1.combine_first(df2)
result
```

```python
# Reshaping with Hierarchical Indexing
# stack()
stacked_left = left.stack()
stacked_left
```

```python
# unstack()
unstacked_left = stacked_left.unstack()
unstacked_left
```
